```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
import numpy as np
plt.style.use('ggplot')
import matplotlib
import matplotlib.pyplot as plt
from pylab import *  
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus']=False

names = [0.0,0.2,0.4,0.6,0.8]  # 刻度值命名
x = [0.0,0.2,0.4,0.6,0.8]    # 横坐标

bleu1 = [0.251,0.255,0.255,0.251,0.245]    # 纵坐标
bleu2 = [0.028,0.027,0.026,0.026,0.025]    # 纵坐标
distinct1 = [0.321,0.308,0.308,0.317,0.342]
distinct2 = [0.747,0.770,0.771,0.784,0.806]
bertscore= [0.600,0.595,0.594,0.593,0.590]

f, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, 1, sharex=True)  # 绘制两个子图
plt.subplots_adjust(wspace=0., hspace=0.35) # 设置 子图间距

ax1.plot(x, bleu1, color='deepskyblue', marker='o', linestyle='solid',label=u'BLEU-1')   # 绘制折线
ax1.set_ylim(0.244, 0.260)  # 设置纵坐标范围
ax2.plot(x, bleu2, color='deepskyblue', marker='o', linestyle='solid',label=u'BLEU-2') # 起图例作用
ax2.set_ylim(0.024, 0.03) # 设置纵坐标范围
ax3.plot(x, distinct1, color='limegreen', marker='o', linestyle='solid',label=u'Distinct-1') # 起图例作用
ax3.set_ylim(0.305, 0.345) # 设置纵坐标范围
ax4.plot(x, distinct2, color='limegreen', marker='o', linestyle='solid',label=u'Distinct-2') # 起图例作用
ax4.set_ylim(0.738, 0.818) # 设置纵坐标范围
ax5.plot(x, bertscore, color='m', marker='o', linestyle='solid',label=u'BERTScore') # 起图例作用
ax5.set_ylim(0.588, 0.603) # 设置纵坐标范围

ax1.grid(axis='both',linestyle='-.')   # 打开网格线
ax2.grid(axis='both',linestyle='-.') # 打开网格线
ax3.grid(axis='both',linestyle='-.')   # 打开网格线
ax4.grid(axis='both',linestyle='-.')   # 打开网格线
ax5.grid(axis='both',linestyle='-.')   # 打开网格线

ax1.grid()
ax2.grid()
ax3.grid()
ax4.grid()
ax5.grid()
plt.xticks(x, names, rotation=0) # 刻度值
# ax1.xaxis.set_major_locator(plt.NullLocator()) # 删除坐标轴的刻度显示

ax1.legend() # 让图例生效
ax2.legend() # 让图例生效
ax3.legend() # 让图例生效
ax4.legend() # 让图例生效
ax5.legend() # 让图例生效
plt.xlabel("Drop Rate") #X轴标签
plt.ylabel("Values", position=(0,3.3)) #Y轴标签

ax1.spines['top'].set_visible(False)    # 边框控制
ax1.spines['bottom'].set_visible(True) # 边框控制
ax1.spines['right'].set_visible(False)  # 边框控制

ax2.spines['top'].set_visible(False)   # 边框控制
ax2.spines['bottom'].set_visible(False) # 边框控制
ax2.spines['right'].set_visible(False)  # 边框控制

ax3.spines['top'].set_visible(False)   # 边框控制
ax3.spines['bottom'].set_visible(False) # 边框控制
ax3.spines['right'].set_visible(False)  # 边框控制

ax4.spines['top'].set_visible(False)   # 边框控制
ax4.spines['bottom'].set_visible(False) # 边框控制
ax4.spines['right'].set_visible(False)  # 边框控制

ax5.spines['top'].set_visible(False)   # 边框控制
ax5.spines['bottom'].set_visible(False) # 边框控制
ax5.spines['right'].set_visible(False)  # 边框控制

# ax1.tick_params(labeltop='off')  
# ax2.tick_params(labeltop='off')  
# ax3.tick_params(labeltop='off')  
# ax4.tick_params(labeltop='off')  
# ax5.tick_params(labeltop='off')  

# 绘制图的数值
for i in range(len(bleu1)):
    ax1.text(x[i], bleu1[i]+0.001, '%s' %round(bleu1[i], 3), ha='center', fontsize=8)
for i in range(len(bleu2)):
    ax2.text(x[i], bleu2[i]-0.002, '%s' %round(bleu2[i], 3), ha='center', fontsize=8)
for i in range(len(distinct1)):
    ax3.text(x[i], distinct1[i]-0.01, '%s' %round(distinct1[i], 3), ha='center', fontsize=8)
for i in range(len(distinct2)):
    ax4.text(x[i], distinct2[i]-0.02, '%s' %round(distinct2[i], 3), ha='center', fontsize=8)
for i in range(len(bertscore)):
    ax5.text(x[i], bertscore[i]-0.004, '%s' %round(bertscore[i], 3), ha='center', fontsize=8)

# # 绘制断层线
# d = 0.  # 断层线的大小
# kwargs = dict(transform=ax1.transAxes, color='k', clip_on=True)
# ax1.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal

# kwargs.update(transform=ax.transAxes, color='k')  # switch to the bottom axes
# ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal

plt.plot()
plt.savefig('ablation_drop.pdf', bbox_inches='tight')

# plt.show()
```
