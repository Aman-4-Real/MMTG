```python
import pickle
```

```python
f = open("datasets/new_data_rating/idx_topics_50.pkl", "rb")
idx_topics = pickle.load(f)
f.close()
f = open("datasets/new_data_rating/final_test_153.pkl", "rb")
final_test_153 = pickle.load(f)
f.close()
```

```python
with open("src_wo_text/res/labels_wo_text.txt", "r") as f:
    labels = f.readlines()
labels = [l.strip() for l in labels]
with open("src_wo_text/res/samples_wo_text.txt", "r") as f:
    samples = f.readlines()
samples = [l.strip() for l in samples]
```

```python
labels_50 = []
i = 0
while i < len(labels):
    for idx_topic in idx_topics:
        if idx_topic[0] == i//10 and idx_topic[1] == final_test_153[i//10]['topic']:
            labels_50 += labels[i:i+10]
            break
    i += 10
len(labels_50)
```

```python
samples_50 = []
i = 0
while i < len(samples):
    for idx_topic in idx_topics:
        if idx_topic[0] == i//10 and idx_topic[1] == final_test_153[i//10]['topic']:
            samples_50 += samples[i:i+10]
            break
    i += 10
len(samples_50)
```

```python
labels_50[:20], samples_50[:20]
```

```python
with open("src_wo_text/res/labels_wo_text_50.txt", "w") as f:
    for line in labels_50:
        f.write(line+'\n')
with open("src_wo_text/res/samples_wo_text_50.txt", "w") as f:
    for line in samples_50:
        f.write(line+'\n')
```
